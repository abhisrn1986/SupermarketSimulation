#### Use the BeautifulSoup object to parse the HTML document 
- You can navigate the html tree down to the tag that contains the data you want.
- There are multiple ways to get to the solution!
- `.find()` always returns the first instance of your "query"
- `.find_all()` returns a list-like object (called a "ResultSet") that contains at least one matching result.
- `.text` returns the actual part of the tag that is outside of the **< angled brackets  >** (i.e. the text)

```python
# find where the poem title and link is in the html page (inspect: class_ item-description-title)
poems_soup.find(class_ = "item-description-title")
```

```python
# use find_all
poems_soup.find_all(class_ = "item-description-title")
```

```python
# split and strip of extracted title
title = poems_soup.find(class_ = "item-description-title").a.text
title
```

```python
title = title.split('\n')
```

```python
title
```

```python
title[11]
```

```python
title[11].strip()
```

```python
titles = []
poemtitle = poems_soup.find_all(class_ = 'item-description-title')
for title in poemtitle:
    title = title.a.text.split('\n')[11].strip()
    titles.append(title)
titles
```

```python
poems_soup.find(class_ = 'item-description-title' ).a['rel']
```

```python
links = []
poemlinks = poems_soup.find_all(class_ = 'item-description-title')
for link in poemlinks:
    links.append(link.a['href'])
links
```

```python
links = []
poemlinks = poems_soup.find_all('div', attrs={'class' : 'item-description'})
for link in poemlinks:
    links.append(link.a['href'])
links
```

```python
import pandas as pd
df = pd.DataFrame({'title':titles, 'link': links})
df
```
